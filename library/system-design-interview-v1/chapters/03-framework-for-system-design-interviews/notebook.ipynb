# Chapter 3: A Framework for System Design Interviews
*System Design Interview*

## TL;DR

System design interviews test **collaboration, communication, and structured thinking** -- not just technical knowledge. Use this 4-step framework:

| Step | Time (45 min) | Focus |
|------|--------------|-------|
| 1. Understand the problem | 3-10 min | Clarify requirements, ask questions |
| 2. High-level design | 10-15 min | Box diagrams, API design, data flow |
| 3. Design deep dive | 10-25 min | Bottlenecks, key components in detail |
| 4. Wrap up | 3-5 min | Bottlenecks, improvements, next steps |

> The final design is less important than the **process** you demonstrate.

## Step 1: Understand the Problem and Establish Design Scope

**DO NOT** jump straight to a solution. Slow down and ask questions.

### Questions to Ask:
- What specific features are we building?
- How many users does the product have?
- How fast does the company anticipate scaling? (3 months, 6 months, 1 year)
- What is the existing technology stack?
- What are the most important features?

### Example: News Feed System

| Question | Answer |
|----------|--------|
| Mobile, web, or both? | Both |
| Most important features? | Make posts, see friends' feed |
| Feed sorted how? | Reverse chronological |
| Max friends per user? | 5,000 |
| Traffic volume? | 10M DAU |
| Media support? | Images and videos |

## Step 2: Propose High-Level Design and Get Buy-in

```mermaid
graph TD
    A[Sketch Component Diagram] --> B[Identify Key Components]
    B --> C[Walk Through Use Cases]
    C --> D[Back-of-Envelope Estimation]
    D --> E[Get Interviewer Buy-in]

    B --> B1[Clients - mobile/web]
    B --> B2[APIs and Web Servers]
    B --> B3[Data Stores]
    B --> B4[Cache and CDN]
    B --> B5[Message Queues]
```

### Key Actions:
1. **Draw box diagrams** with key components on the whiteboard
2. **Walk through concrete use cases** to validate the design
3. **Do back-of-envelope calculations** if scale is important
4. **Collaborate** -- treat the interviewer as a teammate

## Step 3: Design Deep Dive

### Prioritization Strategy

```mermaid
graph TD
    A[Interviewer Signals] --> B{What to focus on?}
    B -->|Senior candidate| C[Performance & Bottlenecks]
    B -->|System focus| D[Specific Component Details]
    B -->|Scale focus| E[Scaling Strategies]
    B -->|General| F[Most Interesting / Novel Parts]

    C --> G[Resource Estimations]
    D --> H[Algorithm Choices]
    E --> I[Sharding / Replication]
    F --> J[Unique Design Decisions]
```

### Rules:
- **Follow interviewer cues** -- they will hint at what interests them
- **Manage your time** -- do not get lost in one component
- **Avoid unnecessary detail** -- e.g., do not explain well-known algorithms in depth
- **Focus on what demonstrates YOUR design skill**

## Step 4: Wrap Up

Use remaining time for:
1. **Identify bottlenecks** and propose improvements
2. **Recap your design** -- refresh the interviewer's memory
3. **Discuss error handling** -- server failures, network loss
4. **Operational concerns** -- monitoring, logging, deployment
5. **Scaling to next level** -- what changes for 10x more users?
6. **Propose refinements** you would make with more time

## Dos and Don'ts

### Dos

| Do | Why |
|----|-----|
| Ask clarifying questions | Avoids designing the wrong system |
| Understand requirements first | No right answer without clear scope |
| Communicate your thinking | The process matters more than the result |
| Suggest multiple approaches | Shows breadth of knowledge |
| Design critical components first | Time is limited |
| Collaborate with the interviewer | They are your teammate |

### Don'ts

| Don't | Why |
|-------|-----|
| Jump into solution immediately | Biggest red flag |
| Go deep on one component too early | Lose the big picture |
| Think in silence | Interviewer cannot evaluate what they cannot see |
| Over-engineer | Shows poor judgment on trade-offs |
| Assume you are done | Keep going until interviewer stops you |
| Be stubborn | Inability to take feedback is a red flag |

## Red Flags to Avoid

| Red Flag | What It Signals |
|----------|----------------|
| Over-engineering | Poor cost/benefit judgment |
| Narrow-mindedness | Cannot consider alternatives |
| Stubbornness | Hard to work with |
| No questions asked | Does not understand requirements gathering |
| Jumping to solution | Poor problem-solving process |
| Silence | Cannot collaborate |

## Takeaways

1. **The process matters more than the final design** -- show structured thinking
2. **Always start by clarifying requirements** -- never assume
3. **Collaborate** -- treat it as a working session, not an exam
4. **Manage time** -- allocate minutes to each step before starting
5. **Communicate constantly** -- think out loud, explain trade-offs
6. **Stay flexible** -- adjust based on interviewer feedback
